### **Задание 1**
Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

 [Нормальное распределение](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html#scipy.stats.norm)  
 Ссылка для scipy.stats.norm

```python
mu = [-2.0, 0.0, 4.0, 4.0, 5.0]
sigma = [2.0, 0.5, 0.5, 1.0, 0.25]

x = np.linspace(-10,10,100)
plt.figure(figsize=(10,8))

# зададим нормально распределенную случайную величину
for i in range(len(mu)):
    norm_rv = sts.norm(loc=mu[i], scale=sigma[i])
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf)

plt.legend(['среднее -2.0: дисперсия 2.0',  'среднее 0.0: дисперсия 0.5', 'среднее 4.0: дисперсия 0.5', 
            'среднее 4.0: дисперсия 1.0', 'среднее 5.0: дисперсия 0.25'])
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.title('Плотности вероятности нескольких выборок')
plt.grid(True)
```

Из графиков видно, что в средних значениях мы имеем максимум, который принимает и положительные и отрицательные значения и ноль. При одинаковых значениях дисперсии получаем одинаковое значение максимума, если средние значения разные. Чем меньше значение дисперсии, тем уже функция плотности вероятности нормального распределения, но тем выше её максимума. Значение максимума функции плотности вероятности распределения может быть больше единицы.  
Главное, функция вероятности распределения всегда меньше или равна единицы.

[Экспоненциальное распределение](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html#scipy.stats.expon)  
Ссылка для scipy.stats.expon

```python
lam = [6.0, 0.5, 1.0, 2.0, 5.0]
pts = [0.0, 1.0, 1.0, 3.0, 3.0]

x = np.linspace(0.0, 5.0, 100)
plt.figure(figsize=(10,8))

# зададим экспоненциально распределенную случайную величину
for i in range(len(lam)):
    expon_rv = sts.expon(loc = pts[i],  scale = 1.0/lam[i])
    pdf = expon_rv.pdf(x)
    plt.plot(x, pdf)

plt.legend(['сдвиг 0.0: параметр 6.0', 'сдвиг 1.0: параметр 0.5', 'сдвиг 1.0: параметр 1.0', 
            'сдвиг 3.0: параметр 2.0', 'сдвиг 3.0: параметр 5.0'])
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.title('Плотности вероятности нескольких выборок')
plt.xlim(0,5)
plt.ylim(0,7)
plt.grid(True)
```

Мы не можем делать отрицательный сдвиг. Чем больше параметр $\lambda$, тем быстрее спадает функция. В начальной точке функция распределения равна параметру $\lambda$.

[Стьюдента распределение](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.t.html#scipy.stats.t)  
Ссылка для scipy.stats.t

```python
sc_ = [0.5, 0.3, 1.0, 2.0, 2.0]
pts = [-1.0, 1.0, 1.0, 2.0, 2.0]
df_ = [1.0, 2.0, 5.0, 1.0, 10.0]

x = np.linspace(-4.0, 6.0, 1000)
plt.figure(figsize=(10,8))

# зададим экспоненциально распределенную случайную величину
for i in range(len(lam)):
    t_rv = sts.t(df = df_[i], loc = pts[i],  scale = sc_[i])
    pdf = t_rv.pdf(x)
    plt.plot(x, pdf)

plt.legend(['df 1.0 loc -1.0 scale 0.5', 'df 2.0 loc 1.0 scale 0.3', 'df 5.0 loc 1.0 scale 1.0', 
            'df 1.0 loc 2.0 scale 2.0', 'df 10.0 loc 2.0 scale 2.0'])
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.title('Плотности вероятности нескольких выборок')
plt.xlim(-4,6)
plt.ylim(0,1.2)
plt.grid(True)
```

Непрерывная случайная величина задана плотностью распределения:$$ f(\xi) = C, x \in [0;5];f(\xi) = 0, x \notin [0;5]$$Найдите C, математическое ожидание $\xi$ а также вероятность попадания $\xi$ в отрезок $[3.5;7]$, т.е. $P(3.5<\xi<7)$.

[Примеры по LaTex'у](https://ru.overleaf.com/learn/latex/Integrals,_sums_and_limits)  
Интегралы, суммы, пределы в LaTex и не только.

$$ f(x) = C, x \in [0;5];f(x) = 0, x \notin [0;5]$$  


Для любой плотности распределения выполняется следующее условие:
$$\int\limits^{+\infty}_{-\infty} f(x)\,dx=1$$  

У нас равномерное распределение на отрезке $[0;5]$.
Следовательно проинтегрировав нашу функцию можно получить, что $C = 1 / 5.$  

Математическое ожидание:  
$$Mx=\int\limits^{+\infty}_{-\infty} f(x)x\,dx$$  
$$Mx=\frac{1}{5}\int\limits^{5}_{0} x\,dx$$  
$$Mx=\frac{1}{5}\frac{x^2}{2}\bigg|_0^5 = 2.5$$

Вполне очевидный ответ, для равномерного распределения. Математическое ожидание оказалось "посередине отрезка".

Вероятность попадания в отрезок $$[3.5;7]$$ равна 
$$P(x \in [3,5;7])=\int\limits^{7}_{3.5} f(x)\,dx=1$$  

$$P = 0.3$$

### **Задание 3**
Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

Количество информации определяется по [формуле Шеннона](https://www.yaklass.ru/p/informatika/10-klass/informatciia-i-informatcionnye-protcessy-11955/izmerenie-informatcii-11900/re-41060398-a186-4359-825b-550b5f0aced1).

У нас три события значит будет три слагаемых в формуле.  
Рассмотрим первый случай: сфетофор горит 54 секунды красным, 4 секунды жёлтым, 28 секунд зелёным.
Нас интересует факт, что светофор горит каким-либо цветом в первой задаче.
Вероятности событий зависит от суммарного времени (цикла светофора), от величины (54 + 4 + 28) = 86 секунд:  
Горит красным: $p_1$ = 54/86  
Горит жёлтым:  $p_2$ =  4/86  
Горит зелёным: $p_3$ = 28/86  
Тогда можно и посчитать количество информации о цвете светофора в текущий момент с помощью кода в следующей ячейке.

```python
t_r = 54
t_y = 4
t_g = 28

t_sum = t_r + t_y + t_g

p_r = t_r / t_sum
p_y = t_y / t_sum
p_g = t_g / t_sum

print(' Вероятность светофор горит красным:', p_r, '', '\n',  
      'Вероятность светофор горит жёлтым:', p_y, '\n',  
      'Вероятность светофор горит зелёным:', p_g)

I_r = -p_r*np.log2(p_r)
I_y = -p_y*np.log2(p_y)
I_g = -p_g*np.log2(p_g)

I = I_r + I_y + I_g

print('\n')
print(' Количество информации светофор горит красным:', I_r, 'бит', '\n',  
      'Количество информации светофор горит жёлтым:', I_y, 'бит', '\n',  
      'Количество информации светофор горит зелёным:', I_g, 'бит', '\n\n',  
      'Количество информации светофор горит:', I, 'бит')
```
